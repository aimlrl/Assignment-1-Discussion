```python
l = [1,2,25,36,31,17,61,19,58,39,67,81]
```

```python
prime = []

even = []

for i in l:
    
    if i % 2 == 0:
        
        if i == 2:
            prime.append(i)
    
        even.append(i)    
    else:
        
        if i == 1:
            prime.append(i)
            
        for divisor in range(2,i):

            if i % divisor == 0:
                break

        if divisor == (i - 1):
            prime.append(i)
            
d = dict(zip(["prime","even"],[prime,even]))

print(d)
```
